## Scrape the following information for each row of the dataset, and save it into new columns in your dataframe.

- The business name
- Owner/operator
- Phone number
- License status (Active, Expired, Etc)
- Physical address

It's basically what we did before, but using the function a little differently.

- *TIP: Same as above, but you'll be returning a `pd.Series` and the `.apply` line is going to be a lot longer*
- *TIP: Save it to a new dataframe!*
- *TIP: Make sure you change your `df` variable names correctly if you're cutting and pasting - there are a few so it can get tricky*

```python
import time 

def get_data(row):
    
    driver.get(row['url'])
    
    company_page = driver.find_elements_by_tag_name('tr')

    #find things
    name_dirty = company_page[4].find_elements_by_tag_name('td')[0].text
    owner_dirty = company_page[5].find_elements_by_tag_name('td')[0].text
    phone_dirty = company_page[6].find_elements_by_tag_name('td')[0].text
    status_dirty = company_page[7].find_elements_by_tag_name('td')[1].text
    address_dirty = company_page[8].find_elements_by_tag_name('td')[1].text   

    #clear things by splitting
    name = name_dirty.split(':   ')[-1]
    owner = owner_dirty.split(':   ')[-1]
    phone = phone_dirty.split(':   ')[-1]
    status = status_dirty.split(':  ')[-1]
    address = address_dirty.split(":")[-1]

    return pd.Series({
            'name': name,
            'owner': owner,
            'phone': phone,
            'status': status,
            'address': address
        })
```

```python
data_df = urls_df.apply(get_data, axis=1).join(urls_df)
data_df
```

```python
#Save as the same filename as before
def get_more_trucks(row):
    
    driver.get(row['url'])
    
    #company_page = driver.find_elements_by_tag_name('tr')

    #find things
    name_dirty = driver.find_element_by_xpath('//*[@id="t1"]/tbody/tr/td/font/table[2]/tbody/tr[2]/td[1]').text
    owner_dirty = driver.find_element_by_xpath('//*[@id="t1"]/tbody/tr/td/font/table[2]/tbody/tr[3]/td[1]').text
    phone_dirty = driver.find_element_by_xpath('//*[@id="t1"]/tbody/tr/td/font/table[2]/tbody/tr[4]/td[1]').text
    status_dirty = driver.find_element_by_xpath('//*[@id="t1"]/tbody/tr/td/font/table[3]/tbody/tr[1]/td[2]').text
    address_dirty = driver.find_element_by_xpath('//*[@id="t1"]/tbody/tr/td/font/table[3]/tbody/tr[2]/td[2]').text   

    #clear things by splitting
    name = name_dirty.split(':   ')[-1]
    owner = owner_dirty.split(':   ')[-1]
    phone = phone_dirty.split(':   ')[-1]
    status = status_dirty.split(':  ')[-1]
    address = address_dirty.split(":")[-1].replace('\n','')

    return pd.Series({
            'name': name,
            'owner': owner,
            'phone': phone,
            'status': status,
            'address': address
        })
```

```python
final_trucks_df = more_urls.apply(get_more_trucks, axis=1).join(more_urls)
final_trucks_df
```
